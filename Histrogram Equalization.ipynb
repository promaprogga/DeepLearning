```python
import cv2
import random
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
originalImage=cv2.imread("/content/dog.jpg") 
image = cv2.resize(originalImage, (400, 300))
cv2_imshow(image)
```

```python
## code to plot histogram
from matplotlib import pyplot as plt
hist,bins = np.histogram(originalImage.flatten(),256,[0,256])
plt.hist(originalImage.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.show()
```

```python
R, G, B = cv2.split(originalImage)

output1_R = cv2.equalizeHist(R)
output1_G = cv2.equalizeHist(G)
output1_B = cv2.equalizeHist(B)

equ = cv2.merge((output1_R, output1_G, output1_B))
EquImage = cv2.resize(equ, (400, 300))
cv2_imshow(EquImage)
```

```python
## code to plot histogram
from matplotlib import pyplot as plt
hist,bins = np.histogram(equ.flatten(),256,[0,256])
plt.hist(equ.flatten(),256,[0,256], color = 'g')
plt.xlim([0,256])
plt.show()
```
