```python
import cv2
import random
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image_path=cv2.imread("/content/flower.jpeg") 
image = cv2. resize(image_path, (400, 300))
cv2_imshow(image)
```

```python
def Gaussian_Noise(image_path):
  x, y, channel= image_path.shape
  # print(channel)
  mean=0
  var=10000
  sigma=var**0.5 
  gauss = np.random.normal(mean,sigma,(x,y,channel))
  gauss = gauss.reshape(x,y,channel)
  noisy = image_path + gauss
  return noisy

img=Gaussian_Noise(image_path)
noisyImg=cv2. resize(img, (400, 300))
cv2_imshow(noisyImg)
```

```python
def recreate(image_path):
  TotalImg=Gaussian_Noise(image_path)
  total=100
  for i in range(total):
    TotalImg= TotalImg+ Gaussian_Noise(image_path)
  avg= TotalImg/total
  return avg

orginalimg= recreate(image_path)
Img=cv2. resize(orginalimg, (400, 300))
cv2_imshow(Img)
```
